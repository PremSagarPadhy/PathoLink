```python
import os

label_dir = r"G:\FRP 2026\alpha-dent\AlphaDent\caries.v3i.yolov11\test\labels"

for file in os.listdir(label_dir):
    if file.endswith(".txt"):
        path = os.path.join(label_dir, file)
        
        with open(path, "r") as f:
            lines = f.readlines()
        
        new_lines = []
        for line in lines:
            parts = line.strip().split()
            
            # Skip empty lines
            if not parts:
                continue 
            
            cls = int(parts[0])
            
            # Merge classes 0 and 1 into class 3
            if cls in [0, 1]:
                parts[0] = "3"
            else:
                parts[0] = str(cls)
            
            new_lines.append(" ".join(parts))
        
        # Check if we have any valid labels to write
        if not new_lines:
            # If the list is empty, the file had no valid data. Delete it.
            os.remove(path)
            print(f"Removed empty label file: {file}")
        else:
            # Otherwise, write the updated labels back to the file
            with open(path, "w") as f:
                f.write("\n".join(new_lines) + "\n")
```

```python
import os
from collections import defaultdict

# Update these paths to your actual label folders
label_folders = {
    "Train": r"G:\FRP 2026\alpha-dent\AlphaDent\caries.v3i.yolov11\train\labels",
    "Valid": r"G:\FRP 2026\alpha-dent\AlphaDent\caries.v3i.yolov11\valid\labels",
    "Test": r"G:\FRP 2026\alpha-dent\AlphaDent\caries.v3i.yolov11\test\labels"
}

for split, path in label_folders.items():
    print(f"--- Statistics for {split} ---")
    
    img_containing_class = defaultdict(int)
    total_annotations = defaultdict(int)
    
    if not os.path.exists(path):
        print(f"Path not found: {path}")
        continue

    label_files = [f for f in os.listdir(path) if f.endswith('.txt')]

    for label_file in label_files:
        with open(os.path.join(path, label_file), 'r') as f:
            lines = f.readlines()
            
            # Get unique classes in this specific image
            classes_in_file = set()
            for line in lines:
                parts = line.split()
                if parts:
                    cls = parts[0]
                    classes_in_file.add(cls)
                    total_annotations[cls] += 1
            
            # For each unique class found, increment the image counter
            for cls in classes_in_file:
                img_containing_class[cls] += 1

    for cls in sorted(total_annotations.keys()):
        print(f"Class {cls}:")
        print(f"  - Images Containing: {img_containing_class[cls]}")
        print(f"  - Annotated Masks/Boxes: {total_annotations[cls]}")
    print("\n")
```
